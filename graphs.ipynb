```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('results.xlsx', index_col=0)
```

```python
data
```

```python
# Your data
tp = data.index
bucket_1 = data['bucket1']
bucket_2 = data['bucket2']
bucket_3 = data['bucket3']
bucket_4 = data['bucket4']

# Create a Matplotlib figure
plt.figure(figsize=(10, 6))

# Set the width of the bars
bar_width = 0.2

# Create the bar positions for each category
bar1 = list(range(len(tp)))
bar2 = [x + bar_width for x in bar1]
bar3 = [x + bar_width for x in bar2]
bar4 = [x + bar_width for x in bar3]

# Create the bar chart
plt.bar(bar1, bucket_1, width=bar_width, label='bucket 1')
plt.bar(bar2, bucket_2, width=bar_width, label='bucket 2')
plt.bar(bar3, bucket_3, width=bar_width, label='bucket 3')
plt.bar(bar4, bucket_4, width=bar_width, label='bucket 4')

# Set labels and title
plt.xlabel('Time Partition')
plt.ylabel('Value')
plt.title('Bar Chart of buckets counting')

# Set x-axis ticks and labels
plt.xticks(
    [x + bar_width * 1.5 for x in range(len(tp))],
    ['01/2008 - 09/2019', '10/2019 - 03/2020', '04/2020 - 12/2020', '01/2021 - 2023']
)

# Labels for buckets
for i, value in enumerate(bucket_1):
    plt.text(bar1[i], value + 0.01, str(round(value, 2)), ha='center', va='bottom', fontsize=9)
for i, value in enumerate(bucket_2):
    plt.text(bar2[i], value + 0.01, str(round(value, 2)), ha='center', va='bottom', fontsize=9)
for i, value in enumerate(bucket_3):
    plt.text(bar3[i], value + 0.01, str(round(value, 2)), ha='center', va='bottom', fontsize=9)
for i, value in enumerate(bucket_4):
    plt.text(bar4[i], value + 0.01, str(round(value, 2)), ha='center', va='bottom', fontsize=9)

# Add a legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
# Your data for all tp
total_bucket_1 = data['avcu_bucket 1'].sum()
total_bucket_2 = data['avcu_bucket 2'].sum()
total_bucket_3 = data['avcu_bucket 3'].sum()
total_bucket_4 = data['avcu_bucket 4'].sum()

# Create a list of total values
total_values = [total_bucket_1, total_bucket_2, total_bucket_3, total_bucket_4]

# Labels for the pie chart
labels = ['bucket 1', 'bucket 2', 'bucket 3', 'bucket 4']

# Explode a slice if you want to highlight it
explode = (0.1, 0.1, 0.1, 0.1)

# Colors for the pie chart
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']

# Create a Matplotlib figure
plt.figure(figsize=(8, 8))

# Create the pie chart
plt.pie(total_values, labels=labels, autopct='%1.1f%%', startangle=140, colors=colors, explode=explode)

# Set title
plt.title('Distribution of avicuerpo bucket Values')

# Add a legend
plt.legend(labels, loc='upper left')

# Show the plot
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
